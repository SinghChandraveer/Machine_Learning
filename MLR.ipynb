```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

df= pd.read_csv("50_Startups.csv")

real_x= df.iloc[:,0:4].values
real_y= df.iloc[:,-1].values

real_x[0:5,:]
```

```python
le= LabelEncoder()
real_x[:,3]= le.fit_transform(real_x[:,3])
real_x[0:5,:]
```

```python
ohe= OneHotEncoder( categorical_features=[3])
real_x= ohe.fit_transform(real_x).toarray()
real_x[0:5,:]
```

```python
train_x,test_x,train_y,test_y= train_test_split(real_x,real_y,test_size=0.2,random_state=0)
mlr= LinearRegression()
mlr.fit(train_x,train_y)
pred_y= mlr.predict(test_x)
pred_y
```
