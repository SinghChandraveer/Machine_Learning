```python
import pandas as pd
import numpy as np

df= pd.read_csv('train_and_test2.csv')
df.head()
```

```python
df.columns
```

```python
df2=df.loc[:,['Passengerid', 'Age', 'Fare', 'Sex','Pclass', 'sibsp','Embarked','2urvived']].astype(float)
df2.columns=['Passengerid', 'Age', 'Fare', 'Sex','Pclass', 'sibsp','Embarked','Survived']
df2.dropna().head()
```

```python
df.dropna(inplace=True)
df2.isnull().sum()
```

```python
x=df2.iloc[:,1:7].values
y=df2.iloc[:,7].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
lr= LogisticRegression(random_state=0)
lr.fit(x_train,y_train)
y_pred= lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
pred_output=lr.predict(x)
pred_output
```

```python
accuracy_score(y, pred_output)
```

```python
output=pd.concat([df['Passengerid'],pd.Series(pred_output).astype(int)],axis=1,sort=False).dropna()
output.columns=['Passengerid','Survived']
output.to_csv('Titanic_analysis.csv')
```
