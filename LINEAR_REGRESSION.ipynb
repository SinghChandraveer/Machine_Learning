```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

a=[1.1,1.3,1.5,2,2.2,2.9,3,3.2,3.2,3.7,3.9,4,4,4.1,4.5,4.9,5.1,5.3,5.9,6,6.8,7.1,7.9,8.2,8.7,9,9.5,9.6,10.3,10.5]
b=[39343,46205,37731,43525,39891,56642,60150,54445,64445,57189,63218,55794,56957,57081,61111,67938,66029,83088,81363,93940,91738,98273,101302,113812,109431,105582,116969,112635,122391,121872]
df1 = pd.DataFrame(list(zip(a,b)), 
               columns =['y.exp', 'salary'])
real_x=df1.iloc[:,0].values.reshape(-1,1)
real_y=df1.iloc[:,1].values.reshape(-1,1)

from sklearn.model_selection import train_test_split
training_x,testing_x,training_y,testing_y= train_test_split(real_x,real_y,test_size=0.3,random_state=0)

LR= LinearRegression()
LR.fit(training_x,training_y)
pred_y= LR.predict(testing_x)
```

```python
plt.scatter(training_x,training_y,color='green')
plt.plot(training_x,LR.predict(training_x),color='blue')
plt.title('sal vs exp training plot')
plt.xlabel('exp')
plt.ylabel('salary')
plt.show
```

```python
plt.scatter(testing_x,testing_y,color='green')
plt.plot(training_x,LR.predict(training_x),color='blue')
plt.title('sal vs exp testing plot')
plt.xlabel('exp')
plt.ylabel('salary')
plt.show
```
